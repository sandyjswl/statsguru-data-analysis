```python
def formatDate(date):
    year = date.split(" ")[-1]
    return year
```

```python
matches_list = data_frame.index.values.tolist()
runs=0
did_not_bat = ["TDNB","DNB","-"]
matches = 0
print(matches_list)
```

```python
def get_runs_for_inns(run):
    if "*" in run:
        run = run.replace("*","")
        return (True,int(run))
    elif run in did_not_bat:
        return (True,0)
    else:
        return (False,int(run))
```

```python
def get_runs_for_match(row):
    total = 0
    inns = 0
    bat1 = row.Bat1
    bat2 = row.Bat2
    first = get_runs_for_inns(bat1)
    second = get_runs_for_inns(bat2)
    if first[0]:
        total = total + first[1]
    else:
        total = total + first[1]
        inns = inns + 1
    if second[0]:
        total = total + second[1]
    else:
        total = total + second[1]
        inns = inns + 1
    return [total,inns]    
        
```

```python
total_runs = 0
total_inns = 0

for i in matches_list:
    current_row = data_frame.loc[i]
    match_data = get_runs_for_match(current_row)
    total_runs+=match_data[0]
    total_inns+=match_data[1]
    #     data_frame.loc[i,"Averages"] = runs/matches
    #     print(data_frame.loc[i,"Runs"])
    data_frame.loc[i, "Average"] = total_runs if total_inns == 0 else round(total_runs / total_inns, 2)
    data_frame.loc[i, "totalRuns"] = total_runs
# print(data_frame)
# return data_frame
```
