```python
import json
with open("result.json") as json_file:
    data = json.load(json_file)
```

```python
import pandas as pd
```

```python
data_frame = pd.DataFrame.from_dict(data,"index")
```

```python
data_frame
```

```python
bat_pos_df = data_frame.groupby(['Pos']).agg({'Runs': 'sum',  'Opposition': 'count'})
```

```python
bat_pos_df
```

```python
bat_pos_df.reset_index(level=0, inplace=True)
```

```python
bat_pos_df.rename(columns={"Opposition":"Matches"},inplace=True)
bat_pos_df
```

```python
bat_pos_df.to_dict("records")
```

```python
import matplotlib.pyplot as plt
import matplotlib
```

```python
plt.figure(figsize=(20,20))
plt.figure().suptitle('test title', fontsize=20)
bat_pos_df.plot(kind='bar',x="Pos",y='Runs',color='Blue',title="Rohit")
plt.savefig("file.jpg",dpi=200)
plt.show
```
