```python
import json
with open("result.json") as json_file:
    data = json.load(json_file)
```

```python
data
```

```python
import pandas as pd
```

```python
data_frame = pd.DataFrame.from_dict(data,"index")
```

```python
data_frame
```

```python
#Runs by year
```

```python
summary_df = data_frame.groupby(['Year']).agg({'Runs': 'sum',  'Opposition': 'count'})
```

```python
summary_df
```

```python
summary_df['Average'] = summary_df["Runs"]/summary_df["Opposition"]
```

```python
summary_df.Average =  summary_df.Average.round(2)
```

```python
summary_df.rename(columns= {"Opposition":"Matches"},inplace=True)
```

```python
summary_df
```

```python
import matplotlib.pyplot as plt
import matplotlib
```

```python
summary_df.reset_index(level=0, inplace=True)
player_name = "Rohit"
```

```python
plt.figure(figsize=(20,20))
plt.figure().suptitle('test title', fontsize=20)
summary_df.plot(kind='bar',x="Year",y='Average',color='Blue',title=player_name)
plt.savefig("file.jpg",dpi=200)
plt.show
```
