```python
from enum import Enum
class Formats(Enum):
    Tests = "Tests"
    ODIs = "ODIs"
    T20Is = "T20Is"

```

```python
import pandas as pd



class StatsAnalysis:
    
    def calculate_average_per_year(self, data_frame):
        total_runs = 0
        matches = 0
        for index, row in data_frame.iterrows():
            current_row_run = row['Runs']
            if '*' in current_row_run:
                run = current_row_run.replace("*", '')
                total_runs = total_runs + int(run)
            else:
                matches += 1
                total_runs = total_runs + int(current_row_run)
        return total_runs if matches == 0 else round((total_runs / matches), 2)

    def formatDate(self, date):
        year = date.split(" ")[-1]
        return year

    def formatRuns(self, runs):
        runs = runs.replace("*", '')
        try:
            return int(runs)
        except:
            return 0

    def clean_data_frame(self, data_frame, format: Formats):
        if format == Formats.Tests:
            data_frame['Bat1'] = data_frame['Bat1'].apply(lambda runs: self.formatRuns(runs))
            data_frame['Bat2'] = data_frame['Bat2'].apply(lambda runs: self.formatRuns(runs))
        data_frame['Runs'] = data_frame['Runs'].apply(lambda runs: self.formatRuns(runs))
        data_frame['Year'] = data_frame['Start Date'].apply(lambda date: self.formatDate(date))
        # data_frame.set_index("Match No", inplace=True)
        return data_frame

    def get_average_by_year_summary_df(self, batting_history, format_enum: Formats):
        data_frame = pd.DataFrame.from_dict(batting_history, orient='index')
        # print(data_frame)
        data_frame = data_frame[data_frame.Runs != "DNB"]
        data_frame = data_frame[data_frame.Runs != "TDNB"]
        # Remove for test matches create a new function
        data_frame = data_frame[data_frame.Runs != "-"]
        data_frame['Year'] = data_frame['Start Date'].apply(lambda date: self.formatDate(date))

        summary_df = data_frame.groupby(['Year'])
        averages = []
        for key, item in summary_df:
            averages.append(self.calculate_average_per_year(summary_df.get_group(key)))

        data_frame["Runs"] = data_frame["Runs"].apply(lambda runs: self.formatRuns(runs))
        summary_df = data_frame.groupby(['Year']).agg({'Runs': 'sum', 'Opposition': 'count'})
        summary_df["Average"] = pd.Series(averages).values

        summary_df.rename(columns={"Opposition": "Matches"}, inplace=True)
        return summary_df

    def create_runs_by_year_data_frame(self, stats_dict, format_enum: Formats):
        data_frame = pd.DataFrame.from_dict(stats_dict, orient="index")
        print(data_frame)
        summary_df = None
        data_frame = self.clean_data_frame(data_frame, format_enum)
        if format_enum == Formats.Tests:
            summary_df = data_frame.groupby(['Year']).agg({'Bat1': 'sum', 'Bat2': 'sum', 'Opposition': 'count'})
        else:
            summary_df = data_frame.groupby(['Year']).agg({'Runs': 'sum', 'Opposition': 'count'})
        summary_df = summary_df.rename(columns={'Opposition': 'Matches'})
        summary_df.reset_index(level=0, inplace=True)
        return summary_df

    def create_average_per_game_dataframe(self, stats_dict):
        data_frame = pd.DataFrame.from_dict(stats_dict, orient="index")
        print(data_frame.head())
        data_frame["Runs"] = data_frame["Runs"].replace("DNB", "None")
        data_frame["Runs"] = data_frame["Runs"].replace("TDNB", "None")
        data_frame["Runs"] = data_frame["Runs"].replace("-", "None")
        data_frame['Average'] = 0
        data_frame['totalRuns'] = 0
        runs = 0
        matches = 0
        # print(data_frame.head())
        matches_list = data_frame.index.values.tolist()
        # print(matches_list)
        for i in matches_list:
            current_row = data_frame.loc[i]
            if current_row.Runs == "None":
                runs = runs + 0
            elif '*' in current_row.Runs:
                run = current_row.Runs.replace("*", "")
                runs = runs + int(run)
            else:
                matches += 1
                runs = runs + int(current_row.Runs)
            #     data_frame.loc[i,"Averages"] = runs/matches
            #     print(data_frame.loc[i,"Runs"])
            data_frame.loc[i, "Average"] = runs if matches == 0 else round(runs / matches, 2)
            data_frame.loc[i, "totalRuns"] = runs
        # print(data_frame)
        return data_frame
```
